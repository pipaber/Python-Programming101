Pregunta 1:  1. Carga diabetes.csv en un DataFrame df_diabetes. 2. Muestra las 5 primeras filas. 3. Imprime el número total de filas, columnas y el tipo de cada columna. 4. Convierte la columna Outcome a tipo categórico y verifica el cambio imprimiendo la información del DataFrame (df_diabetes.info()).

```python
import pandas as pd

df_diabetes = pd.read_csv('diabetes.csv')
df_diabetes.head()
```

```python
print(df_diabetes.shape)
print(df_diabetes.dtypes)
```

```python
df_diabetes["Outcome"] = df_diabetes["Outcome"].astype("category")
df_diabetes.info()
```

Pregunta 2:  1. Crea un DataFrame df_filtrado con todas las filas de df_diabetes donde Glucose > 120. 2. Agrega a df_filtrado una columna BMI_Category que clasifique el BMI así: ○ "Bajo" si BMI < 18.5 ○ "Normal" si 18.5 ≤ BMI < 25 ○ "Sobrepeso" si 25 ≤ BMI < 30 ○ "Obeso" si BMI ≥ 30 3. Muestra las primeras 5 filas de df_filtrado con la nueva columna incluida.

```python
df_filtrado = df_diabetes.copy()[df_diabetes["Glucose"] > 120]
df_filtrado
```

```python
def bmi_category(bmi):
    if bmi < 18.5:
        return "Bajo"
    elif 18.5 <= bmi < 25:
        return "Normal"
    elif 25 <= bmi < 30:
        return "Sobrepeso"
    else:
        return "Obeso"

df_filtrado["BMI_Category"] = df_filtrado["BMI"].apply(bmi_category)
df_filtrado.head()
```

Pregunta 3:  1. Carga age_groups.csv en un DataFrame df_age_groups. 2. Haz un merge de df_diabetes con df_age_groups por la columna Age. El resultado se llamará df_combinado. Debe ser un left merge para mantener todas las filas de df_diabetes. 3. Muestra las 5 primeras filas de df_combinado para verificar que la columna AgeGroup se haya agregado correctamente.

```python
df_age_groups = pd.read_csv('age_groups.csv')
df_combinado = pd.merge(df_diabetes, df_age_groups, on='Age', how='left')
df_combinado.head()
```

Pregunta 4:  1. Agrupa df_combinado por AgeGroup. 2. Calcula el promedio de Glucose y el promedio de BMI para cada grupo de edad. 3. Ordena el resultado por el promedio de Glucose de mayor a menor. 4. Muestra el DataFrame resultante.

```python
(df_combinado
    .groupby("AgeGroup")
    .agg({"Glucose": "mean", "BMI": "mean"})
    .sort_values(by="Glucose", ascending=False)
)
```

Pregunta 5:  1. A partir de df_combinado, filtra las filas donde AgeGroup = "Adulto" y Outcome = 1 (casos positivos de diabetes en adultos). 2. De este subconjunto, selecciona los 5 registros con el mayor valor de BMI. 3. Muestra únicamente las columnas Age, Glucose, BMI, DiabetesPedigreeFunction, ordenadas por BMI de mayor a menor.

```python
df_combinado_filtered = df_combinado[(df_combinado["AgeGroup"]=="Adulto") & (df_combinado["Outcome"]==1)]
df_combinado_filtered.nlargest(5, "BMI")[["Age", "Glucose", "BMI", "DiabetesPedigreeFunction"]].sort_values(by="BMI", ascending=False)
```

Pregunta 6:  1. Usando df_combinado, crea una tabla pivote que tenga: ○ Índice: AgeGroup ○ Columnas: Outcome ○ Valores: promedio de Glucose y BMI 2. La pivot table debe mostrar el promedio de Glucose y BMI por combinación de AgeGroup y Outcome. 3. De la tabla resultante, filtra y muestra únicamente las filas (AgeGroups) donde el promedio de BMI para cualquiera de los valores de Outcome sea mayor a 30.

```python
df_pivot = df_combinado.pivot_table(index="AgeGroup", columns="Outcome", values=["Glucose", "BMI"], aggfunc="mean")
df_pivot
```

```python
df_pivot[df_pivot["BMI", 1] >30]
```
